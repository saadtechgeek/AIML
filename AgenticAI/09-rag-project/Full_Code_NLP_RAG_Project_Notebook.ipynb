### Data Chunking

```python
# Initialize a text splitter that breaks documents into token-based chunks
# using the cl100k_base tokenizer (GPT-compatible encoding).
# Each chunk holds up to 512 tokens, with a 50-token overlap
# between consecutive chunks to preserve context across boundaries.
text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    encoding_name='cl100k_base',  # Tokenizer encoding for accurate token counting
    chunk_size=512,               # Maximum number of tokens per chunk
    chunk_overlap=50              # Overlapping tokens between chunks to retain context
)
```

```python
# Load the PDF document and split it into smaller chunks
# using the previously defined text_splitter.
# The result is a list of document chunks, each containing
# up to 512 tokens with 50-token overlap between them.
document_chunks = load_file.load_and_split(text_splitter)
```

### System and User Prompt Template

```python
qna_system_message = """You are a helpful medical AI assistant that answers questions accurately using information from the Merck Manual.
Your objective is to provide structured, reliable, and easy-to-understand responses grounded in the provided context.
Guidelines to follow:
- Base all answers strictly on the provided context and medical facts
- Break down complex medical or technical topics into clear, digestible explanations
- Use bullet points or key points to organize responses when needed
- Acknowledge gaps in knowledge openly rather than fabricating information
- Maintain a professional, empathetic, and informative tone throughout"""
```

```python
qna_user_message_template = """###Source Information
{context}

###Query
{question}

###Answer
Drawing from the source information above, deliver a clear, accurate, and informative response to the query."""
```

### Response Function

```python
def generate_rag_response(user_input, k=3, max_tokens=128, temperature=0, top_p=0.95, top_k=50):
    global qna_system_message, qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = retriever.invoke(input=user_input)

    context_list = [d.page_content for d in relevant_document_chunks]

    # Combine document chunks into a single context
    context_for_query = ". ".join(context_list)

    user_message = qna_user_message_template.replace('{context}', context_for_query)
    user_message = user_message.replace('{question}', user_input)

    prompt = qna_system_message + '\n' + user_message

    # Generate the response
    try:
        response = llm(
                  prompt=prompt,
                  max_tokens=max_tokens,
                  temperature=temperature,
                  top_p=top_p,
                  top_k=top_k
                  )

        # Extract and print the model's response
        response = response['choices'][0]['text'].strip()
    except Exception as e:
        response = f'Sorry, I encountered the following error: \n {e}'

    return response
```

## Output Evaluation

Let us now use the LLM-as-a-judge method to check the quality of the RAG system on two parameters - retrieval and generation. We illustrate this evaluation based on the answeres generated to the question from the previous section.

- We are using the same Mistral model for evaluation, so basically here the llm is rating itself on how well he has performed in the task.

```python
groundedness_rater_system_message = """You are an expert evaluator responsible for assessing how well an AI assistant's response is grounded in the provided context.
Given a question, context, and answer, rate the groundedness of the response on a scale of 1 to 5 using the criteria below.
1 = Fully ungrounded, the answer has no support from the provided context
2 = Largely ungrounded, the answer draws very little from the provided context
3 = Partially grounded, only some portions of the answer are backed by the context
4 = Mostly grounded, the answer is largely supported with only minor unsupported claims
5 = Completely grounded, every part of the answer is fully supported by the provided context

Respond with only a single number (1-5) followed by a concise explanation."""
```

```python
relevance_rater_system_message = """You are a skilled evaluator responsible for assessing how relevant an AI assistant's response is to the given question.
Given a question, context, and answer, rate the relevance of the response on a scale of 1 to 5 using the criteria below.
1 = Completely irrelevant, the answer does not address the question in any way
2 = Largely irrelevant, the answer barely touches on what was asked
3 = Partially relevant, the answer addresses some but not all aspects of the question
4 = Mostly relevant, the answer addresses the question well with only minor gaps
5 = Fully relevant, the answer directly and thoroughly addresses all aspects of the question

Respond with only a single number (1-5) followed by a concise explanation."""
```

```python
user_message_template = """
###Question
{question}

###Context
{context}

###Answer
{answer}
"""
```

```python
def generate_ground_relevance_response(user_input, k=3, max_tokens=128, temperature=0, top_p=0.95, top_k=50):
    global qna_system_message, qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = retriever.invoke(input=user_input)
    context_list = [d.page_content for d in relevant_document_chunks]
    context_for_query = ". ".join(context_list)

    # Combine user_prompt and system_message to create the prompt
    prompt = f"""[INST]{qna_system_message}\n
                {'user'}: {qna_user_message_template.format(context=context_for_query, question=user_input)}
                [/INST]"""

    response = llm(
            prompt=prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            echo=False
            )

    answer = response["choices"][0]["text"]

    # Combine user_prompt and system_message to create the prompt
    groundedness_prompt = f"""[INST]{groundedness_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    # Combine user_prompt and system_message to create the prompt
    relevance_prompt = f"""[INST]{relevance_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    response_1 = llm(
            prompt=groundedness_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            echo=False
            )

    response_2 = llm(
            prompt=relevance_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            echo=False
            )

    return response_1['choices'][0]['text'], response_2['choices'][0]['text']
```

## Actionable Insights and Business Recommendations

**Actionable Insights**

1. RAG Significantly Outperforms Base LLM
The notebook demonstrates that plain LLM responses are generic and based solely on training data, while RAG responses are grounded in the Merck Manual, a trusted 4,000+ page medical reference. This confirms that RAG is essential for domain-specific accuracy in healthcare applications.

2. Prompt Engineering Matters
Responses with a well-defined system prompt covering role, tone, and structure are consistently more reliable and organized than unprompted LLM outputs. The quality of the system message directly impacts clinical usefulness.

3. LLM-as-a-Judge Reveals Quality Gaps
Using Mistral to rate its own responses on groundedness and relevance helps identify weak points in the retrieval pipeline, particularly when retrieved chunks do not contain sufficient context to answer the query accurately.

4. Context Window is a Real Constraint
The 2300-token limit caused failures during evaluation. For medical queries involving long context, this is a hard bottleneck that requires architectural attention and optimization.

5. Quantized Models Are Viable for Prototyping
Mistral-7B Q4/Q6 performed adequately on medical QA tasks, demonstrating that cost-efficient quantized models can power healthcare prototypes without the need for enterprise-grade hardware.

**Business Recommendations**

1. Deploy as a Clinical Decision Support Tool
Integrate the RAG system into hospital EHR workflows so physicians can query the medical manual in real time, reducing time spent on manual literature lookup during diagnosis.

2. Expand the Knowledge Base
The Merck Manual is a strong foundation, but adding specialty-specific sources such as oncology protocols, ICU guidelines, and drug interaction databases would make the system more comprehensive and clinically valuable.

3. Upgrade Context Window for Production
Replace Mistral-7B with a model supporting 8K to 32K context such as Mistral-7B-v0.3 or LLaMA 3 to handle complex multi-part medical queries without token overflow errors.

4. Implement Human-in-the-Loop Validation
The LLM-as-a-judge approach is a good starting point, but production systems should include clinician review for flagged low-groundedness responses to prevent incorrect medical guidance from reaching end users.

5. Ensure Regulatory Compliance
Before deployment, ensure the system complies with HIPAA data privacy standards and FDA AI/ML guidance for clinical decision support. Clear disclaimers should also be included stating that the system is an assistive tool and not a replacement for physician judgment.
